```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
```

```python
df = pd.read_csv('data.csv')
df['correct'] = (df.true == df.pred)
df['(true=1,pred=1)'] = (df.true == 1) &  (df.pred == 1)
df['(true=1,pred=0)'] = (df.true == 1) &  (df.pred == 0)
df['(true=0,pred=1)'] = (df.true == 0) &  (df.pred == 1)
df['(true=0,pred=0)'] = (df.true == 0) &  (df.pred == 0)
df.head()
```

```python
recent = df[(df.id >= 1500) & (df.id <= 10000)]
user_df = recent[(recent.age == 0) & (recent.gender == 'other')]
user_df
```

```python
print (user_df.groupby(['expt_id','duration']).id.count())
print (user_df.groupby(['expt_id','duration']).correct.sum())
```

```python
results = df.groupby(['expt_id','duration']).agg({
    'correct': ['sum'],
    '(true=1,pred=1)': ['sum'],
    '(true=1,pred=0)': ['sum'],
    '(true=0,pred=1)': ['sum'],
    '(true=0,pred=0)': ['sum'],
    'true': ['sum'],
    'id': ['count']
})

results['accuracy'] = results.correct['sum']/results.id['count']
results['(pred=1|true=1)'] = results['(true=1,pred=1)']['sum']/results['true']['sum']
results['(pred=0|true=1)'] = results['(true=1,pred=0)']['sum']/results['true']['sum']
results['(pred=1|true=0)'] = results['(true=0,pred=1)']['sum']/(results['id']['count'] - results['true']['sum'])
results['(pred=0|true=0)'] = results['(true=0,pred=0)']['sum']/(results['id']['count'] - results['true']['sum'])


results = results[[
    'accuracy',
    '(pred=1|true=1)',
    '(pred=0|true=1)',
    '(pred=1|true=0)',
    '(pred=0|true=0)',
]].reset_index()


gender_results = df.groupby(['expt_id','duration', 'gender']).agg({
    'correct': ['sum'],
    '(true=1,pred=1)': ['sum'],
    '(true=1,pred=0)': ['sum'],
    '(true=0,pred=1)': ['sum'],
    '(true=0,pred=0)': ['sum'],
    'true': ['sum'],
    'id': ['count']
})

gender_results['accuracy'] = gender_results.correct['sum']/gender_results.id['count']
gender_results['(pred=1|true=1)'] = gender_results['(true=1,pred=1)']['sum']/gender_results['true']['sum']
gender_results['(pred=0|true=1)'] = gender_results['(true=1,pred=0)']['sum']/gender_results['true']['sum']
gender_results['(pred=1|true=0)'] = gender_results['(true=0,pred=1)']['sum']/(gender_results['id']['count'] - gender_results['true']['sum'])
gender_results['(pred=0|true=0)'] = gender_results['(true=0,pred=0)']['sum']/(gender_results['id']['count'] - gender_results['true']['sum'])


gender_results = gender_results[[
    'accuracy',
    '(pred=1|true=1)',
    '(pred=0|true=1)',
    '(pred=1|true=0)',
    '(pred=0|true=0)',
]].reset_index()

gender_results[gender_results.gender=='female']
```

```python
plt.rcParams.update({'font.size': 14})

fig = plt.figure(figsize=(10,12))
ax1 = fig.add_subplot(211)
ax1.set_title("Face Classification Overall Accuracy vs. Time Shown")
ax1.set_xlabel("Time Shown (ms)")
ax1.set_ylabel("Accuracy = P(true == pred)")
ax2 = fig.add_subplot(212)
ax2.set_title("Fake Image FPR vs. Time Shown")
ax2.set_xlabel("Time Shown (ms)")
ax2.set_ylabel("FPR = P(pred=1|true=0)")

id_map = {1:("Exp 1 (real vs GAN)", "m", "-"), 2:("Exp 2 (real vs GAN; no eyes)", "m", "--")}

for expt_id, (label, color, linestyle)  in id_map.items():
    exp_data = results[results.expt_id == expt_id]
    ax1.plot(exp_data.duration, exp_data.accuracy, color=color, linestyle=linestyle, marker='o', label=label)
    ax2.plot(exp_data.duration, exp_data['(pred=1|true=0)'], color=color, linestyle=linestyle, marker='o', label=label)

ax1.axhline(y=0.5, color="k", linestyle="--", label="Random Guessing = 0.5")
ax2.axhline(y=0.5, color="k", linestyle="--", label="Random Guessing = 0.5")

ax1.legend(loc=0)
ax2.legend(loc=0)

fig.savefig('./plots/overall_accuracy.svg')
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(10,12))
ax1 = fig.add_subplot(211)
ax1.set_title("Face Classification Overall Accuracy vs. Time Shown")
ax1.set_xlabel("Time Shown (ms)")
ax1.set_ylabel("Accuracy = P(true == pred)")

ax2 = fig.add_subplot(212)
ax2.set_title("Fake Image FPR vs. Time Shown")
ax2.set_xlabel("Time Shown (ms)")
ax2.set_ylabel("FPR = P(pred=1|true=0)")

gender_id_map = {
    ('male', 1):("Male; Exp 1 (real vs GAN)", "b", "-"), 
    ('male', 2):("Male; Exp 2 (real vs GAN; no eyes)", "b", "--"),
    ('female', 1):("Female; Exp 1 (real vs GAN)", "r", "-"), 
    ('female', 2):("Female; Exp 2 (real vs GAN; no eyes)", "r", "--")
}

for (gender, expt_id), (label, color, linestyle)  in gender_id_map.items():
    exp_data = gender_results[(gender_results.gender == gender) & (gender_results.expt_id == expt_id)]
    ax1.plot(exp_data.duration, exp_data.accuracy, color=color, linestyle=linestyle, marker='o', label=label)
    ax2.plot(exp_data.duration, exp_data['(pred=1|true=0)'], color=color, linestyle=linestyle, marker='o', label=label)

ax1.axhline(y=0.5, color="k", linestyle="--", label="Random Guessing = 0.5")
ax2.axhline(y=0.5, color="k", linestyle="--", label="Random Guessing = 0.5")

ax1.legend(loc=0)
ax2.legend(loc=0)

plt.tight_layout()
plt.show()
```

```python
img_df = df.groupby('filename').agg({
    'id': 'count',
    'expt_id':'max',
    'correct': 'sum',
    'true': 'max'
}).reset_index()
img_df['accuracy'] = img_df.correct / img_df.id



gender_img_df = df.groupby(['filename', 'gender']).agg({
    'id': 'count',
    'expt_id':'max',
    'correct': 'sum',
    'true': 'max'
}).reset_index()
gender_img_df['accuracy'] = gender_img_df.correct / gender_img_df.id
```

```python
N = 6
true = 0
expt_id = 1
ascending = True

data = img_df[(img_df.true == true) & (img_df.expt_id == expt_id)].sort_values(by=['accuracy', 'id'], ascending=ascending).reset_index()

fig = plt.figure(figsize=(24,6))
for i in range(N):
    ax = fig.add_subplot('%d%d%d' % (1,N,i+1))
    ax.set_title("%s\nattempts=%d, correct=%d" % (data.filename[i], data.id[i], int(data.correct[i])))
    filepath = os.path.join('./server/static', data.filename[i])
    ax.imshow(mpimg.imread(filepath))
    ax.set_yticklabels([])
    ax.set_xticklabels([])

fig.savefig('./plots/hard_fake_exp1.svg')
plt.show()
```

```python
N = 6
true = 1
expt_id = 1
ascending = True
gender = 'female'

data = gender_img_df[(gender_img_df.true == true) & (gender_img_df.expt_id == expt_id) & (gender_img_df.gender == gender)].sort_values(by=['accuracy', 'id'], ascending=ascending).reset_index()

fig = plt.figure(figsize=(20,4))
for i in range(N):
    ax = fig.add_subplot('%d%d%d' % (1,N,i+1))
    ax.set_title("%s\nattempts=%d, correct=%d" % (data.filename[i], data.id[i], int(data.correct[i])))
    filepath = os.path.join('./server/static', data.filename[i])
    ax.imshow(mpimg.imread(filepath))

plt.show()
```
